## 2.- Preprocesamiento

### A.- Division del dataset

```python
x = df.drop('diagnosis', axis=1)
y = df['diagnosis']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

### B.- Transformacion

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train = pd.DataFrame(scaler.fit_transform(x_train), columns=x_train.columns)
x_test = pd.DataFrame(scaler.transform(x_test), columns=x_test.columns) 
```

```python
from sklearn.preprocessing import LabelEncoder

lab_encoder = LabelEncoder()
y_train = lab_encoder.fit_transform(y_train)
y_test = lab_encoder.transform(y_test)
```

### C.- Seleccion de caracteristicas

```python
from sklearn.ensemble import RandomForestClassifier

tree_model = RandomForestClassifier()

tree_model.fit(x_train, y_train)
```

```python
tree_model.feature_importances_
```

```python
df_importancias = pd.DataFrame({
    'feature': list(x_train), 
    'importance': tree_model.feature_importances_
    }).sort_values(by='importance',ascending=False)
```

```python
df_importancias
```

```python
sns.barplot(data=df_importancias, x='importance', y='feature')
```

```python
df_importancias['feature'][:11]
```

```python
x_train = x_train[df_importancias['feature'][:11]]
x_test= x_test[df_importancias['feature'][:11]]
```

## 3.- Entrenamiento

### A.- Seleccion del modelo

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB


models = {
    'lr_model': LogisticRegression(),
    'tree': DecisionTreeClassifier(),
    'rf_model': RandomForestClassifier(),
    'svc': SVC(),
    'naive bayes': GaussianNB(),
}
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

resultado =  []

for name, model in models.items():
    model.fit(x_train, y_train)

    y_pred = model.predict(x_test)

    acccuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    resultado.append({'model': name, 'acurracy': acccuracy, 'precision': precision, 'recall': recall, 'f1_score': f1})
```

```python
metricas_df = pd.DataFrame(resultado)
metricas_df
```

```python
metrics = ['acurracy', 'precision', 'recall','f1_score']

plt.figure(figsize=(25, 4))
for i, metric in enumerate(metrics, start=1):
    plt.subplot(1, 4, i)
    sns.barplot(data=metricas_df, x='model', y=metric)
```

### B.- Busqueda de hiperparametros

```python
hiperparametros = {
    'penalty': ['l1', 'l2', 'elasticnet', 'none'],  
    'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'], 
    'max_iter': [100, 200, 300, 400, 500, 1000, 2000],         
    'tol': [1e-4, 1e-3, 1e-2, 1e-1],                
    'class_weight': [None, 'balanced']              
}
```

```python
from sklearn.model_selection import GridSearchCV

lr_model = GridSearchCV(estimator=LogisticRegression(), param_grid=hiperparametros, cv=5)

lr_model.fit(x_train, y_train)
```

```python
lr_model.best_params_
```

## 4.- Evaluacion

```python
y_pred = lr_model.predict(x_test)

print(
    'acurracy: ', accuracy_score(y_test, y_pred),
    '\nprecision: ', precision_score(y_test, y_pred),
    '\nrecall: ', recall_score(y_test, y_pred),
    '\nf1 score', f1_score(y_test, y_pred) 
)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

sns.heatmap(cm, annot=True, xticklabels=['c. 0', 'c. 1',], yticklabels=['c. 0', 'c. 1'])
plt.xlabel("PredicciÃ³n")
plt.ylabel("Real")
```
