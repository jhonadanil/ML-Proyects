## 2.- Preprocesamiento

### A.- Division del dataset

```python
x = df.drop('Class', axis=1)
y = df['Class']

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

### B.- Seleccion de caracteristicas

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier()

rf_model.fit(x_train, y_train)
```

```python
rf_model.feature_importances_
```

```python
imp_df = pd.DataFrame({
    'feature': list(x_train.columns),
    'importance': rf_model.feature_importances_
}).sort_values(by='importance', ascending=False)
```

```python
sns.barplot(data=imp_df, y='feature', x='importance')
```

```python
x_train_red = x_train[imp_df['feature'][:9]]
x_test_red= x_test[imp_df['feature'][:9]]
```

## 3.- Entrenamiento

```python
from sklearn.ensemble import IsolationForest

if_model = IsolationForest(contamination=0.1) 
if_model.fit(x_train)
```

## 4.- Evaluacion 

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

y_pred = if_model.predict(x_test)
y_pred = np.where(y_pred == -1, 1, 0)
```

```python
print(
    'acurracy: ', accuracy_score(y_test, y_pred),
    '\nprecision: ', precision_score(y_test, y_pred),
    '\nrecall: ', recall_score(y_test, y_pred),
    '\nf1 score', f1_score(y_test, y_pred) 
)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

sns.heatmap(cm, annot=True, xticklabels=['c. 0', 'c. 1',], yticklabels=['c. 0', 'c. 1'])
plt.xlabel("PredicciÃ³n")
plt.ylabel("Real")
```
